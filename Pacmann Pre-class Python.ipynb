````python
# import pandas and model_selection module of scikit-learn
import pandas as pd
from sklearn import model_selection

# Training data is in a CSV file called train.csv
df = pd.read_csv("train.csv")

# we create a new column called kfold and fill it with -1
df["kfold"] = -1

# the next step is to randomize the rows of the data
df = df.sample(frac=1).reset_index(drop=True)

# initiate the kfold class from model_selection module
kf = model_selection.KFold(n_splits=5)

# fill the new kfold column
for fold, (trn_, val_) in enumerate(kf.split(X=df)):
    df.loc[val_, 'kfold'] = fold

# save the new csv with kfold column
df.to_csv("train_folds.csv", index=False)
````

**Order of Operation (PEMDAS)**

Setiap operasi matematika memiliki **urutan dalam penyelesaiannya**. Jika dalam satu ekspresi terdapat lebih dari satu operasi, maka harus di selesaikan berdasarkan operasi yang lebih tinggi prioritasnya. Untuk lebih mudah mengingat aturan ini bisa menggunakan akronim **PEMDAS**:

| Operator                                       | Simbol |
|------------------------------------------------|--------|
| **P**arentheses (Tanda Kurung)                  | ( )    |
| **E**xponents (Pangkat)                   | **     |
| **M**ultiplication, **D**ivision (kali, bagi, pembagian (dengan output pembulatan kebawah), hasil sisa pembagian)       | * , / , //, % |
| **A**ddition and **S**ubtraction (Tambah, Kurang)                                          | + , -      |

Jika operasi memiliki tingkat prioritas yang sama maka dikerjakan mulai dari kiri

Untuk tipe data `string`, caranya juga berlaku seperti cara sebelumnya


```python

```

**Limiting Float Decimal Point**

Jika kita berhadapan dengan tipe data float dengan angka di belakang koma yang banyak, kita bisa menyederhanakan jumlah angkanya menjadi yang kita inginkan.

Cara pertama bisa menggunakan method `round()`. Method ini akan membulatkan bilangan dibelakang koma menjadi yang kita inginkan.


contoh:

```python
numb_1 = 14.33733
round_num = round(num_1, 2) # 2 disini merupakan jumlah angka di belakang koma
print(round_num)
output=14.34
```

Jika menggunakan f-string, kita bisa lebih mudah menggunakan sintaks (`.2f`). Sintaks tersebut akan membulatkan dua angka di belakang koma. Angka 2 tersebut bisa anda ganti dengan berapapun angka yang kita inginkan

Contoh:
```python
numb_2 = 14.33733
print(f"angka = {numb_2: .2f}")
output = 14.34
```

````python
# Karena ambulance dan pemadan dikecualikan
# simpan ambulance dan pemadam kebakaran pada suatu tuple
exc = ('Ambulance', 'Pemadam Kebakaran')

# buat perulangan dengan kondisi
# cek apakah plat tersebut ganjil atau genap dan
# pastikan lewat jalur yang benar
# jika tidak lewat jalur yang benar proses tilang

for plat, jenis, jalur in record:
    param = plat[-4]

    if jalur in exc:
        print('{} dengan plat nomor {} | tidak di proses e-Tilang'.format(jalur,plat))

    elif jenis in exc:
        print('{} dengan plat nomor {} | tidak di proses e-Tilang'.format(jenis,plat))


    elif int(param) % 2 == 1:

        if jalur == 'Jalur A':
            print('{} Plat Nomor {} masuk {} | Tidak di Proses e_Tilang'.format(jenis,plat,jalur))

        else:
            print('{} Plat Nomor {} masuk {} | Proses e_Tilang'.format(jenis,plat,jalur))

    elif int(param) % 2 == 0:

        if jalur == 'Jalur B':
            print('{} Plat Nomor {} masuk {} | Tidak di Proses e_Tilang'.format(jenis,plat,jalur))

        else:
            print('{} Plat Nomor {} masuk {} | Proses e_Tilang'.format(jenis,plat,jalur))


````

````python
# buat fungsi
gaji = 4_500_000
uang_lembur = 50_000
uang_lembur2 = 30_000
max = 20

def income_employee():

    status = input('Kategori pegawai (Tetap / Kontrak) : ')
    jam_lembur = input('Berapa jam lembur : ')
    lembur = int(jam_lembur)

    if status == 'Kontrak' or status == 'kontrak':

      if lembur <= max:
        lemburan = (uang_lembur2 * lembur)
      else:
        lemburan = (uang_lembur2 * 20)

      pendapatan = (4500000 - (4500000 * 0.7)) + lemburan

    elif status == 'Tetap' or status == 'tetap':

      if lembur <= max:
        lemburan = (uang_lembur * lembur)
      else:
        lemburan = (uang_lembur * 20)

      pendapatan = gaji + lemburan

    else:
      print('Anda salah input')

    pajak = 0.1 * pendapatan
    print('\nTotal uang lembur = {} '.format(lemburan))
    print('Total pendapatan karyawan = {} '.format(pendapatan))

    net_income = pendapatan - pajak
    print('\nPajak yang harus dibayarkan = {} '.format(pajak))
    print('Total pendapatan karyawan setelah dipotong pajak = {} '.format(net_income))



````

````python
list1 = [10, 8, 6, 4, 2, 0, -2, -4, 4, 55, 123, 0, 0, 0]

def ratio_list(list_angka):
    pos, neg, zero = 0,0,0

    for num in list_angka:
        if num > 0:
            pos += 1
        elif num < 0:
            neg += 1
        else:
            zero += 1
    
    rasio_pos = pos / len(list_angka)
    rasio_neg = neg / len(list_angka)
    rasio_zero = zero / len(list_angka)

    return rasio_pos, rasio_neg, rasio_zero

ratio_list(list1)
````
